```python
import matplotlib.pyplot as plt
```

```python
x = [2,4,5,6,7,11,15]
y = ['a','b','c','d','e','f','g']
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'o')
```

```python
plt.plot(x,y,'g')
```

```python
# ms = marker size
# ls = line style
plt.plot(x,y,'o',c='y',ms=7,ls='--') 
```

```python
plt.grid()
```

```python
plt.plot()
```

```python
plt.plot(x,y)
plt.grid()
```

```python
plt.plot(x,y)
plt.grid()
plt.title('Makka Mishi', color='green',fontsize=15)
```

```python
plt.plot(x,y,'o')
plt.grid()
plt.title('Makka Mishi', c='g',fontsize=15)
plt.xlabel('No of Makka')
plt.ylabel('No of Mishi')
```

```python
plt.plot(x,y,'o','o','o')
plt.grid()
plt.title('Makka Mishi', c='g',fontsize=15)
plt.xlabel('No of Makka')
plt.ylabel('No of Mishi')
```

```python
plt.plot(x,y,'o','o','o','o','o','o','o','o','o','o','o','o','o','o','o')
plt.grid()
plt.title('Makka Mishi', c='g',fontsize=15)
plt.xlabel('No of Makka')
plt.ylabel('No of Mishi')
```

```python

plt.plot(x,y,'o',ls="-")
plt.grid()
plt.title('Makka Mishi', c='g',fontsize=15)
plt.xlabel('No of Makka')
plt.ylabel('No of Mishi')
```
