```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Simple Line Plot")
plt.plot(x,y)
plt.show()
```

### Bar chart

```python
categories = ['A','B','C']
values = [5,7,9]
```

```python
plt.bar(categories, values)
plt.title('Bar Chart')
plt.show()
```

## Scatter plot

```python
x = [1,2,3,4,5]
y = [2,4,1,8,7]
```

```python
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.show()
```

## Sub plots

```python
# Multiple Plots in One Figure

plt.subplot(1,2,1)  #plt.subplot(nrows, ncols, index)
plt.plot(x,y)
plt.title('Line Plot')

plt.subplot(1,2,2)
plt.bar(categories, values)
plt.title('Bar Chart')

plt.tight_layout()  ## Adjust spacing between subplots
plt.show()
```
